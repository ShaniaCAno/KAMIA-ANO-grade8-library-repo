```python
import requests
import firebase_admin
from firebase_admin import credentials, db


github_raw_url = "https://raw.githubusercontent.com/ShaniaCAno/KAMIA-ANO-grade8-library-repo/refs/heads/main/books.json"

response = requests.get(github_raw_url)
book_data = response.json()

if not firebase_admin._apps:
    cred = credentials.Certificate("firebase_key.json")
    firebase_admin.initialize_app(cred, {
        'databaseURL': 'https://kamia-ano-grade8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/'
    })


ref = db.reference('/')
ref.set(book_data)

print("--- Full Database Display ---")
print(ref.get())

print("\n--- Library Inventory Report ---")
total_library_books = 0

for book in book_data['books']:
    title = book['title']
    available = book['details']['available']
    copies = book['details']['copies']

    print(f"Book Title: {title} | Available: {available}")

    total_library_books += copies

print(f"\nTotal number of books in the library: {total_library_books}")
```
